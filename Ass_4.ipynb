### Computing Distances with Different types.

```python
import numpy as np
from scipy.spatial.distance import cosine, cityblock, euclidean, minkowski
```

```python
def compute_distance(point1, point2, distance_type, p=2):
    """Compute the distance between two points based on the selected distance type."""
    if distance_type == 'euclidean':
      return euclidean(point1, point2)
    elif distance_type == 'manhattan':
      return cityblock(point1, point2)
    elif distance_type == 'minkowski':
      return minkowski(point1, point2, p)
    elif distance_type == 'cosine':
      return cosine(point1, point2)
    else:
      raise ValueError("Unknown distance type")
```

```python
# Example usage
if __name__ == "__main__":
   # Sample points
   point1 = np.array([1, 2, 3])
   point2 = np.array([4, 5, 6])
```

```python
# User input for distance measure
distance_type = input("Enter distance type (euclidean, manhattan, minkowski, cosine):").strip().lower()

if distance_type == 'minkowski':
   p = float(input("Enter value for p (e.g., 1 for Manhattan, 2 for Euclidean): "))
else:
   p = 2 # Default value for Minkowski if not used
    
# Compute and display distance
distance = compute_distance(point1, point2, distance_type, p)
print(f"{distance_type.capitalize()} distance between {point1} and {point2} is: {distance}")
```
