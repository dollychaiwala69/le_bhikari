### Covariance and Correlation and their Matrices

```python
import numpy as np
import pandas as pd
```

```python
# Function to calculate covariance
def calculate_covariance(X, Y): 
    mean_X = np.mean(X)
    mean_Y = np.mean(Y)
    covariance = np.mean((X - mean_X) * (Y - mean_Y))
    return covariance
```

```python
# Function to calculate correlation
def calculate_correlation(X, Y):
    covariance = calculate_covariance(X, Y)
    std_X = np.std(X)
    std_Y = np.std(Y)
    correlation = covariance / (std_X * std_Y)
    return correlation
```

```python
# Function to compute covariancematrix
def covariance_matrix(data):
    return np.cov(data, rowvar=False)
```

```python
# Function to compute correlation matrix
def correlation_matrix(data):
    return np.corrcoef(data, rowvar=False)
```

```python
# Sample dataset
data = {
'X': [1, 2, 3, 4, 5],
'Y': [2, 4, 6, 8, 10],
'Z': [5, 4, 3, 2, 1]
}
```

```python
# Convert to pandas DataFrame
df = pd.DataFrame(data)
```

```python
# Compute covariance and correlation between 'X' and 'Y' 
cov_xy = calculate_covariance(df['X'],df['Y'])
cor_xy = calculate_correlation(df['X'], df['Y'])
```

```python
# Compute covariance matrix and correlation matrix
cov_matrix = covariance_matrix(df)
corr_matrix = correlation_matrix(df)
```

```python
print("Covariance between X and Y:", cov_xy) 
print("Correlation between X and Y:", cor_xy)
```

```python
print("Covariance Matrix:\n", cov_matrix) 
print("Correlation Matrix:\n", corr_matrix)
```
